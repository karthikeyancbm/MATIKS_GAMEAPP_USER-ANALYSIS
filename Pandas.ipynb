```python
import pandas as pd
student = pd.DataFrame()
student['name'] = ["arti","bhavana","suresh","vijay"]
student['email'] = ["arti@gmail.com","bhav@gmail.com","suresh@gmail.com","vijay@gmail.com"]
student['course'] = ["DS","DE","ALML","DS"]
student


bank = pd.DataFrame()
bank['name'] = ["arti","bhavana","suresh","vijay","ankit","vinay"]
bank['account_number'] = [156151,154265,1552563,155356,1552458,1552859]
bank['balance'] = [10,12,9,8,25,11]
bank
```

```python
# inner joins are used to fetch the data common in both data frames
df = student.merge(bank,how="inner")
df
```

```python
df = student.merge(bank)
df
```
